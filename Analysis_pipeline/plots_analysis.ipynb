```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Results.csv")
data
```

```python
plt.plot(data["B0"],data["delta_r"])
plt.title(r"Damping Factor variation with $\vec{B}$",size=19)
plt.xlabel(r"$B_{0}$",size=18)
plt.ylabel(r"$\delta_{r}$",size=18)
plt.grid(True)
plt.savefig("Images/B0_delta_variation.png", dpi=600)
plt.show()
```

```python
plt.plot(data["B0"],data["Kmin"],color="red",label=r"$k_{min}$")
plt.plot(data["B0"],data["Kmax"],color="blue",label=r"$k_{max}$")
plt.plot(data["B0"],data["Lower_k"],color="orange",label=r"lower k")
plt.title(r"Wavenumber variation with $\vec{B}$",size=19)
plt.xlabel(r"$B_{0}$",size=18)
plt.ylabel(r"k",size=18)
plt.grid(True)
plt.legend(fontsize=16)
plt.savefig("Images/k_variation.png", dpi=600)
plt.show()
```
