```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filepath = "../Helicon_Parametric_Analysis/RESULTS/"
filename = "generic"
name = "B=20G;n=2e13"
```

```python
data0 = pd.read_csv(f"{filepath}Beta_k_relation.csv")
#data0 = data0[ (data0["k"] < 15) & (data0[" beta "] > 0.35) ]
data0 = data0[ (data0["k"] < 15) & (data0[" beta "] > 2.3) ]
data0.columns
```

```python
%matplotlib notebook
plt.plot(data0[" beta "], data0["k"],label=f"{name}")
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r"$\beta (cm^{-1})$")
plt.ylabel(r"$k(cm^{-1})$")
plt.grid(True)
plt.legend()
plt.savefig(f"{filepath}Images/k_beta_{name}.jpg", dpi=600)
plt.show()
```
