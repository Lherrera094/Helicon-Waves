```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataBH = pd.read_csv("../Helicon_Waves_physics/RESULTS/B_H_wave.csv")
dataBTG = pd.read_csv("../Helicon_Waves_physics/RESULTS/B_TG_wave.csv")
dataEH = pd.read_csv("../Helicon_Waves_physics/RESULTS/E_H_wave.csv")
dataETG = pd.read_csv("../Helicon_Waves_physics/RESULTS/E_TG_wave.csv")
```

```python
r = 2.5
dataH["r(m)"] = dataH["r(m)"]/r
dataTG["r(m)"] = dataTG["r(m)"]/r
```

```python
%matplotlib notebook
plt.plot(dataH["r(m)"], dataH[" B_r(T)"],color="red",label="Br")
plt.plot(dataH["r(m)"], dataH[" B_theta(T)"],color="purple",label="Bth")
plt.plot(dataH["r(m)"], dataH[" B_z(T) "],color="darkcyan",label="Bz")
plt.title("H-Wave Magnetic Field")
plt.legend()
plt.xlabel("r/a")
plt.ylabel(r"$\vec{B}$(G)")
plt.show()
plt.savefig("Images/BesselH1.jpg", dpi=600)
```

```python
%matplotlib notebook
plt.plot(dataTG["r(m)"], dataTG[" B_r(T)"],color="red",label="Br")
plt.plot(dataTG["r(m)"], dataTG[" B_theta(T)"],color="purple",label="Bth")
plt.plot(dataTG["r(m)"], dataTG[" B_z(T) "],color="darkcyan",label="Bz")
plt.title("TG-Wave Magnetic Field (G)")
plt.xlabel("r/a")
plt.ylabel(r"$\vec{B}$(G)")
plt.legend()
plt.show()
plt.savefig("Images/BesselTG1.jpg", dpi=600)
```
