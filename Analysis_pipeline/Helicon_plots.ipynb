```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataBH = pd.read_csv("../Helicon_Waves_physics/RESULTS/B_H_wave.csv")
dataBTG = pd.read_csv("../Helicon_Waves_physics/RESULTS/B_TG_wave.csv")
dataEH = pd.read_csv("../Helicon_Waves_physics/RESULTS/E_H_wave.csv")
dataETG = pd.read_csv("../Helicon_Waves_physics/RESULTS/E_TG_wave.csv")
```

```python
r = 2.5
dataBH["r(m)"] = dataBH["r(m)"]/r
dataBTG["r(m)"] = dataBTG["r(m)"]/r
dataEH["r(m)"] = dataEH["r(m)"]/r
dataETG["r(m)"] = dataETG["r(m)"]/r
```

```python
%matplotlib notebook
plt.plot(dataBH["r(m)"], dataBH[" B_r(T)"],color="red",label="Br")
plt.plot(dataBH["r(m)"], dataBH[" B_theta(T)"],color="purple",label="Bth")
plt.plot(dataBH["r(m)"], dataBH[" B_z(T) "],color="darkcyan",label="Bz")
plt.title("H-Wave Magnetic Field")
plt.legend()
plt.xlabel("r/a")
plt.ylabel(r"$\vec{B}$(G)")
plt.show()
plt.savefig("Images/BesselH1.jpg", dpi=600)
```

```python
%matplotlib notebook
plt.plot(dataBTG["r(m)"], dataBTG[" B_r(T)"],color="red",label="Br")
plt.plot(dataBTG["r(m)"], dataBTG[" B_theta(T)"],color="purple",label="Bth")
plt.plot(dataBTG["r(m)"], dataBTG[" B_z(T) "],color="darkcyan",label="Bz")
plt.title("TG-Wave Magnetic Field (G)")
plt.xlabel("r/a")
plt.ylabel(r"$\vec{B}$(G)")
plt.legend()
plt.show()
plt.savefig("Images/BesselTG1.jpg", dpi=600)
```

```python
%matplotlib notebook
plt.plot(dataEH["r(m)"], dataEH[" E_r(T)"],color="red",label="Er")
plt.plot(dataEH["r(m)"], dataEH[" E_theta(T)"],color="purple",label="Eth")
plt.plot(dataEH["r(m)"], dataEH[" E_z(T) "],color="darkcyan",label="Ez")
plt.title("H-Wave Electric Field")
plt.legend()
plt.xlabel("r/a")
plt.ylabel(r"$\vec{E}$(G)")
plt.show()
plt.savefig("Images/H-wEfield.jpg", dpi=600)
```

```python
%matplotlib notebook
plt.plot(dataETG["r(m)"], dataETG[" E_r(T)"],color="red",label="Er")
plt.plot(dataETG["r(m)"], dataETG[" E_theta(T)"],color="purple",label="Eth")
plt.plot(dataETG["r(m)"], dataETG[" E_z(T) "],color="darkcyan",label="Ez")
plt.title("TG-Wave Electric Field")
plt.legend()
plt.xlabel("r/a")
plt.ylabel(r"$\vec{E}$(G)")
plt.show()
plt.savefig("Images/TG-wEfield.jpg", dpi=600)
```
