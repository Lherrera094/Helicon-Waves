```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filepath = "../Helicon_Waves_physics/RESULTS/"
filename = "generic"
```

```python
dataBH = pd.read_csv(f"{filepath}B_H_wave.csv")
dataBTG = pd.read_csv(f"{filepath}B_TG_wave.csv")
dataEH = pd.read_csv(f"{filepath}E_H_wave.csv")
dataETG = pd.read_csv(f"{filepath}E_TG_wave.csv")
```

```python
#normalize radius
r = 2.5
dataBH["r(m)"] = dataBH["r(m)"]/r
dataBTG["r(m)"] = dataBTG["r(m)"]/r
dataEH["r(m)"] = dataEH["r(m)"]/r
dataETG["r(m)"] = dataETG["r(m)"]/r
```

```python
#%matplotlib notebook
plt.plot(dataBH["r(m)"], dataBH[" B_r(T)"],color="red",label="Br")
plt.plot(dataBH["r(m)"], dataBH[" B_theta(T)"],color="darkcyan",label="Bth")
plt.plot(dataBH["r(m)"], dataBH[" B_z(T) "],color="blue",label="Bz")
plt.title("H-Wave Magnetic Field")
plt.legend()
plt.xlabel("r/a")
plt.ylabel(r"$\vec{B}$(G)")
plt.grid(True)
plt.savefig(f"{filepath}Images/H_waveB_{filename}.jpg", dpi=600)
plt.show()
```

```python
#%matplotlib notebook
plt.plot(dataBTG["r(m)"], dataBTG[" B_r(T)"],color="red",label="Br")
plt.plot(dataBTG["r(m)"], dataBTG[" B_theta(T)"],color="darkcyan",label="Bth")
plt.plot(dataBTG["r(m)"], dataBTG[" B_z(T) "],color="blue",label="Bz")
plt.title("TG-Wave Magnetic Field (G)")
plt.xlabel("r/a")
plt.ylabel(r"$\vec{B}$(G)")
plt.legend()
plt.grid(True)
plt.savefig(f"{filepath}Images/TG_waveB_{filename}.jpg", dpi=600)
plt.show()
```

```python
#%matplotlib notebook
plt.plot(dataEH["r(m)"], dataEH[" E_r(T)"],color="red",label="Er")
plt.plot(dataEH["r(m)"], dataEH[" E_theta(T)"],color="darkcyan",label="Eth")
plt.plot(dataEH["r(m)"], dataEH[" E_z(T) "],color="blue",label="Ez")
plt.title("H-Wave Electric Field")
plt.legend()
plt.xlabel("r/a")
plt.ylabel(r"$\vec{E}$(G)")
plt.grid(True)
plt.savefig(f"{filepath}Images/H_waveE_{filename}.jpg", dpi=600)
plt.show()
```
